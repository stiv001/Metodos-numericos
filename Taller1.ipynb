```python
import numpy as np
import matplotlib.pyplot as plt

# Función cúbica
def funcion_cubica(x):
    return x**3 - 7*x**2 + 4*x + 12

coeficientes = [1, 4, -7, 12]

raices = np.roots(coeficientes)

# Rango de valores para x
valores_x = np.linspace(-5, 5, 100)

valores_y = funcion_cubica(valores_x)

# Graficar la función
plt.plot(valores_x, valores_y, label=r"$f(x) = x^3 - 7x^2 + 4x + 12$", color="blue")

# Marcar raíces
for raiz in raices:
    if np.isreal(raiz):
        plt.plot(raiz.real, 0, 'ro')  
        plt.text(raiz.real, 0, f'{raiz.real:.2f}', color="blue", ha='center', va='bottom')

# Ajustar límites
plt.xlim(-5, 5)
plt.ylim(-50, 50)


plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Gráfica de la función cúbica")
plt.axhline(0, color="black", linewidth=0.5)  # Eje y=0
plt.axvline(0, color="black", linewidth=0.5)  # Eje x=0
plt.grid(True)  
plt.legend() 
plt.show()  # Mostrar gráfico

```
